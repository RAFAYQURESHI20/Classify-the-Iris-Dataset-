```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
load=load_iris()
```

```python
dir(load)
```

```python
load.feature_names
```

```python
load.target
```

```python
df=pd.DataFrame(load.data,columns=load.feature_names)
df
```

```python
x=df
```

```python
y=load.target
```

```python
df['target']=load.target
```

```python
df['flower_n']=df.target.apply(lambda x:load.target_names[x])
df
```

```python
df1=df[df.flower_n=='setosa']
df1
```

```python
df2=df[df.flower_n=='versicolor']
df2
```

```python
df3=df[df.flower_n=='virginica']
df3
```

```python
import matplotlib.pyplot as plt
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green',marker='+')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='blue',marker='.')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green',marker='+')
plt.scatter(df2['petal length (cm)'],df3['petal width (cm)'],color='blue',marker='.')
```

```python
x=df.drop(['target','flower_n'],axis='columns')
x
```

```python
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

**SVM**

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model1.predict(x_test)
```

```python
model1.predict([[5.1,3.5,1.4,0.2]])
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,model1.predict(x_test))
cm
```

```python
import seaborn as sn
plt.figure(figsize=(15,10))
sn.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
